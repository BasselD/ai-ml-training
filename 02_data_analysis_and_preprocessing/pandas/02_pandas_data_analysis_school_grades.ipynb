The describe() function in Pandas does not count null values. It only counts non-null values. This means that if a column has any null values, the count statistic in the describe() output will be less than the total number of rows in the DataFrame.

The describe() function also does not count unique values. It only provides summary statistics for numeric data types. For categorical data types, it provides the count of non-null values, the number of unique values, the mode (the most frequently occurring value), and the number of times the mode was observed.

### Aggregation and Plotting

```python
# find the average math score by gender
# use groupy function
dfAvgM = df[['gender','math_score']].groupby('gender').mean()
dfAvgM
```

```python
dfAvgM.plot(kind='bar')
```

```python
import matplotlib.pyplot as plt
#use the index function to get the gender types
plt.bar(dfAvgM.index, dfAvgM['math_score'])
```

```python
#get the lowest writing score by ethnicity group
dfMin = df[['class_group', 'writing_score']].groupby('class_group').min()

dfMin
```

```python
#let's build a pie chart for this aggregation
dfMin.plot(kind='pie', y='writing_score')
```

#### Advanced Aggregation using `agg` and `groupby`

```python
# advanced aggregation using agg with a custom function
def custom_range(x):
    return x.max() - x.min()

df.groupby('parental_level_of_education')['math_score'].agg(custom_range)
```

```python
# specifying multiple functions/calculations
df.groupby('parental_level_of_education')['math_score'].agg(['min', 'max', 'mean'])
```

```python
# advanced 2
df.groupby('parental_level_of_education').agg({'math_score': 'mean', 'reading_score': 'sum'})  # Calculate mean of 'Value' and sum of 'Quantity' per category
```
