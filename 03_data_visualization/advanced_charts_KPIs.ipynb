```python
import matplotlib.pyplot as plt

# Sample data
kpi_data = [
    {'label': 'Revenue', 'value': 120, 'target': 100},
    {'label': 'Cost', 'value': 90, 'target': 80},
    {'label': 'Profit', 'value': 30, 'target': 20}
]

# Create figure
fig, ax = plt.subplots(figsize=(8, 4))

# Plot KPIs
for i, kpi in enumerate(kpi_data):
    ax.text(0.1, 0.8 - i * 0.2, kpi['label'], ha='left', va='center', fontsize=16)
    ax.text(0.4, 0.8 - i * 0.2, f"{kpi['value']:.0f}", ha='center', va='center', fontsize=24, color='blue')
    ax.text(0.6, 0.8 - i * 0.2, f"Target: {kpi['target']:.0f}", ha='left', va='center', fontsize=14, color='gray')

# Customize
ax.axis('off')
ax.set_title('KPI Dashboard', fontsize=20)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Sample data
kpi_data = [
    {'label': 'Revenue', 'value': 120, 'target': 100, 'change': 28, 'color': 'green'},
    {'label': 'Cost', 'value': 90, 'target': 80, 'change': -10, 'color': 'green'},
    {'label': 'Profit', 'value': 30, 'target': 20, 'change': -5, 'color': 'red'},
    {'label': 'MLR', 'value': '86%', 'target': 40, 'change': 25, 'color': 'red'}
]

# Create figure
fig, ax = plt.subplots(figsize=(12, 3), dpi=200)

# Plot KPIs
for i, kpi in enumerate(kpi_data):
    # Rounded rectangle frame
    rect = patches.FancyBboxPatch(
        xy=(0.06 + i * 0.25, 0.1),  # (x, y)
        width=0.18,  # width
        height=0.7,  # height
        boxstyle=f'round,pad=0.01,rounding_size={0.01}',  # adjust roundness
        ec='#cccbca',
        fc='white'
    )
    ax.add_patch(rect)

    # Description
    ax.text(0.06 + i * 0.25, 0.7, kpi['label'], ha='left', va='center', fontsize=14, color='#999999')

    # Value
    ax.text(0.15 + i * 0.25, 0.5, f"{kpi['value']}", ha='center', va='center', fontsize=36, color='#424242',weight='bold')

    # Change
    if kpi['change'] > 0:
        ax.text(0.15 + i * 0.25, 0.35, f"↑ {kpi['change']}%", ha='left', va='center', fontsize=14, color=kpi['color'])
    else:
        ax.text(0.15 + i * 0.25, 0.35, f"↓ {abs(kpi['change'])}%", ha='left', va='center', fontsize=14, color=kpi['color'])

    # Target
    ax.text(0.15 + i * 0.25, 0.2, f"Target: {kpi['target']:.0f}", ha='center', va='center', fontsize=12, color='gray')



# Customize
ax.axis('off')
ax.set_title('KPI Dashboard', fontsize=20)

plt.show()
```
