## Additional Merging Exercise

```python
import pandas as pd
```

```python
data = {'Name': ['Matt', 'John', 'Tammy', 'Bob', 'Nancy'],
        'Age': [33,27,19, 29, 18],
        'City': ['New York', 'Nashville', 'San Diego', 'Asheville','Atlanta'],
        'University': ['NSU', 'Vanderbilt', 'USD', 'UNC', 'GSU'],
        'Class': ['', '', '', '',''],
        }
df1 = pd.DataFrame.from_dict(data)
df1
```

```python
data2 = {
        'Name': ['Matt', 'John', 'Tammy', 'Bob', 'Nancy'],
        'Class': ['Mathematics', 'Statistics', 'History', 'Biology','English'],
        }

df2 = pd.DataFrame.from_dict(data2)
df2
```

```python
df_updated =pd.merge(df1,df2, on="Name")
df_updated
```

```python
df_updated = df_updated.drop(columns='Class_x')
df_updated
```

```python
mask = df1["Name"].isin(df2["Name"])
df1.loc[mask, "Class"] = df2["Class"]
```
